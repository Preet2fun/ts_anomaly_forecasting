# Error-Trend-seasonality(ETS) Decompositions

```python
import numpy as np
import pandas as pd
%matplotlib inline

airline = pd.read_csv('airline_passengers.csv',index_col=0,parse_dates=True)
# we can not perform ETS decompositions on any of misssing data value so drop it using dropna 
airline = airline.dropna()
airline.head()
```

```python
airline.plot(figsize=(12,5))
```

```python
from statsmodels.tsa.seasonal import seasonal_decompose
result = seasonal_decompose(airline['Thousands of Passengers'],model='multiplicative')
```
